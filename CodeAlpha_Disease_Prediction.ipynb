```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import math
from sklearn.svm import SVC
```

```python
medical_data=pd.read_csv("/content/drive/MyDrive/Datasets/disease_prediction/medical_examination (1).csv")
```

```python
medical_data.head()
```

```python
medical_data.loc[(medical_data['gender']==3) & (medical_data['height']>=165), 'gender'] = 1
medical_data.loc[(medical_data['gender']==3) & (medical_data['height']<165), 'gender'] = 2
```

*  Age : int(days)
*  Cholesterol:
      * 1: normal, 2: above normal, 3: well above normal
*   ap_hi  : Systolic blood pressure
*   ap_lo  : Diastolic blood pressure
*   active : Physical activity
*   cardio : Presence or absence of cardiovascular disease
* Glucose : 1: normal, 2: above normal, 3: well above normal
* gendre: 1:men, 2:female


```python
medical_data['age'] = medical_data['age'].apply(lambda x: math.floor(x / 365.25))
```

```python
medical_data.info()
```

## Split our dataset

```python
X_train, X_test= train_test_split(medical_data, test_size=0.2, random_state=42, shuffle=True)
```

```python
Y_train=X_train.pop('cardio')
Y_test=X_test.pop('cardio')
```

## **Choose and train the model**

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,predictions))
```
